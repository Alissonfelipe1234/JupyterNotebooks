```python
cidades = cidades[['CO_MUNICIPIO','NO_MUNICIPIO', 'CO_SIGLA_ESTADO']]
```

```python
cidades = cidades[['CO_MUNICIPIO','NO_MUNICIPIO', 'CO_SIGLA_ESTADO']]
```

```python
cidades = cidades[cidades.CO_SIGLA_ESTADO != 'AZ']
```

```python
cidades = cidades[cidades.CO_SIGLA_ESTADO != 'TE']
```

```python
cidades = cidades[cidades.CO_SIGLA_ESTADO != 'UF']
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
f = open("cidades.sql", "w+", encoding="utf-8")
for row in range(0, cidades['CO_MUNICIPIO'].count()-1):
    codigo = cidades.iloc[row][0]
    cidade = cidades.iloc[row][1]
    sigla = cidades.iloc[row][2]
    print(cidade + " - " + sigla)
    estadoid = estados.loc[estados['CO_SIGLA'] == sigla].values[0][0]
    f.write('INSERT INTO cidades(idCidade, strNomeCidade, strEstadoSigla, rIdEstado) VALUES("{0}","{1}","{2}","{3}");\n'.format(codigo, cidade, sigla, estadoid))
f.close()
```
